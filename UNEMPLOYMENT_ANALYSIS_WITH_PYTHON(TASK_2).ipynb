ESTIMATED NUMBER OF EMPLOYEES ACCORDING TO THE DIFFERENT REGIONS OF INDIA

DASHBOARD TO ANALYZE RATE OF UNEMPLOYMENT IN EACH STATE

```python
unemployment = data[["states","Region","Estimated Unemployment Rate"]]
figure=px.sunburst(unemployment,path=["Region","states"],
                   values="Estimated Unemployment Rate",
                   width=700,height=700,color_continuous_scale="dRY1nG",
                   title="Unemployment Rate in India")
figure.show()
```

ANALYZING DATA BY BAR GRAPHS

```python
fg=px.bar(data,x="states",y="Estimated Unemployment Rate",color="Region",
          title="STATE WISE UNEMPLOYMENT RATE",template="plotly")
fg.update_layout(xaxis={"categoryorder":"total descending"})
fg.show()
```

ANALYZING DATA BY BAR GRAPHS REGION WISE

```python
fg=px.bar(data,x="Region",y="Estimated Unemployment Rate",color="Region",
          title="REGION WISE UNEMPLOYMENT RATE",template="plotly")
fg.update_layout(xaxis={"categoryorder":"total descending"})
fg.show()
```

ANALYZING DATA BY BOX PLOT
